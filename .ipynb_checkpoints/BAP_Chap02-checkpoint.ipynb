グリッドアプローチ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
plt.style.use('seaborn-darkgrid')

def posterior_grid(grid_points=100, heads=6, tosses=9):
    """
    A grid implementation for the coin-flip problem
    """
    # パラメータに関して合理的な区間を定義する
    grid = np.linspace(0, 1, grid_points)
    
    # その区間に点のグリッド（等間隔）を位置付ける
    prior = np.repeat(5, grid_points)
    
    # グリッドの各点に関して、尤度と事前分布を掛け合わせる
    likelihood = stats.binom.pmf(heads, tosses, grid)
    unstd_posterior = likelihood * prior
    posterior = unstd_posterior / unstd_posterior.sum()
    return grid, posterior

points = 15
h, n = 1, 4
grid, posterior = posterior_grid(points, h, n)
plt.plot(grid, posterior, 'o-', label='heads = {}\ntosses = {}'.format(h, n))
plt.xlabel(r'$\theta$', fontsize=14)
plt.legend(loc=0, fontsize=16)
plt.show()
```

モンテカルロ（乱数を使用して何らかの値を見積もる手法）

例）円周率の計算

```python
N = 10000

# 1×1の正方形内にランダムな点を打つ（一様分布に従った乱数を生成）
x, y = np.random.uniform(-1, 1, size=(2, N))

# 原点（左下の頂点）から距離が 1 以下なら 1 ポイント，1 より大きいなら 0 ポイント追加
inside = (x**2 + y**2) <= 1

# 以上の操作を N 回繰り返す，総獲得ポイントを X とするとき，4XN が円周率の近似値になる
# https://note.com/shimakaze_soft/n/n9547f5c0bae0
pi = inside.sum()*4/N
error = abs((pi - np.pi)/pi)*100

outside = np.invert(inside)

plt.plot(x[inside], y[inside], 'b.')
plt.plot(x[outside], y[outside], 'r.')
plt.plot(0, 0, label='$\hat\pi$ = {:4.3f}\nerror = {:4.3f}%'.format(pi, error), alpha=0)
plt.axis('square')
plt.legend(frameon=True, framealpha=0.9, fontsize=16)
plt.show()
print('円周率：' + str(pi))
```

簡単なメトロポリス法を実行する

```python
def metropolis(func, steps=10000):
    """
    A very simple Metropolis implementation
    """
    samples = np.zeros(steps)
    old_x = func.mean()
    old_prob = func.pdf(old_x)
    
    for i in range(steps):
        new_x = old_x + np.random.normal(0, 0.5)
        new_prob = func.pdf(new_x)
        acceptance = new_prob/old_prob
        if acceptance >= np.random.random():
            samples[i] = new_x
            old_x = new_x
            old_prob = new_prob
        else:
            samples[i] = old_x
    return samples

func = stats.beta(0.4, 1)
samples = metropolis(func=func)
x = np.linspace(0.01, 0.09, 100)
y = func.pdf(x)
plt.xlim(0, 1)
plt.plot(x, y, 'r-', lw=3, label='True distribution')
plt.hist(samples, bins=30, normed=True, label='Estimated distribution')
plt.ylabel('$pdf(x)$', fontsize=14)
plt.legend(fontsize=14)
plt.tight_layout()
```

コイン投げ問題の計算的なアプローチ

```python
np.random.seed(123)
n_experiments = 4
theta_real = 0.35
data = stats.bernoulli.rvs(p=theta_real, size=n_experiments)
print(data)
```

```python
import pymc3 as pm
with pm.Model() as our_first_model:
    
    # 事前分布の設定
    theta = pm.Beta('theta', alpha=1, beta=1)
    # 尤度
    y = pm.Bernoulli('y', p=theta, observed=data)
    
    # 最大事後確率（Maximum a Posteriori; MAP）
    start = pm.find_MAP()
    # サンプリング法の定義：メトロポリス-ヘイスティング法
    step = pm.Metropolis()
    
    # サンプリングの実行
    trace = pm.sample(1000, step=step, start=start)
```

```python
# バーンイン（burn-in）
# 有限のサンプルからより良い抽出をするために、最初の方のサンプルを一定数削除する
burnin = 100

# サンプルチェーンを用いてパラメータを変更
chain = trace[burnin:]
pm.traceplot(chain, lines={'theta':theta_real})
```

<h3>パラメータの事後分布を見る前に確認しておくべき事項その１</h3>

<b>MCMCの代表性をチェックする</b>

チェイン内の値は事後分布を代表していなければならない。
チェインの任意の初期値に過度に影響をうけるべきでなく、
一部に留まることなく事後分布の範囲を十分に探索すべきである

確認する指標

<ul>
    <li>見た目によるチェック</li>
    <ol>
        <li>トレースプロット</li>
        <li>確率密度プロット</li>
    </ol>
    <li>数値によるチェック</li>
    <ol>
        <li>Gelman-Rubin統計量</li>
    </ol>
</ul>

事後分布を並列サンプリングし、トレースプロットする

```python
with our_first_model:
    step = pm.Metropolis()
    multi_trace = pm.sample(1000, step=step, cores=4)
    
burnin = 100
multi_chain = multi_trace[burnin:]
pm.traceplot(multi_chain, lines={'theta':theta_real})
```

MCMCは短く言えば確率分布が与えられたときに、その確率分布に従う標本を作る手法です。 「標本→確率分布」ではなく、「確率分布→標本」です。

まずは、なんで「確率分布→標本」の作成が必要なのか、その辺を確認します。 一言で言うと、積分が困難な状況でベイズ推定したい際に積分を標本をもとにした総和で近似して解く場合に有効だからです。
