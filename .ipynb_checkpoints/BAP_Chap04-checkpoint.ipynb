様々なガンマ分布を出力する

```python
import pymc3 as pm
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-darkgrid')
np.set_printoptions(precision=2)
pd.set_option('display.precision', 2)

rates = [1, 2, 5]
scales = [1, 2, 5]

x = np.linspace(0, 20, 100)
f, ax = plt.subplots(len(rates), len(scales), sharex=True, sharey=True)
for i in range(len(rates)):
    for j in range(len(scales)):
        rate = rates[i]
        scale = scales[j]
        rv = stats.gamma(a=rate, scale=scale)
        ax[i, j].plot(x, rv.pdf(x))
        ax[i, j].plot(0, 0, label="$\\alpha$ = {:3.2f}\n$\\theta$ = {:3.2f}".format(rate, scale), alpha=0)
        ax[i, j].legend()
        
ax[2, 1].set_xlabel("$x$")
ax[1, 0].set_ylabel("$pdf(x)$")
plt.show()
```

回帰直線を既知として、人工的にデータを生成する

```python
np.random.seed(314)
# 標本数
N = 100
# α
alfa_real = 2.5
# β
beta_real = 0.9
# 許容誤差
eps_real = np.random.normal(0, 0.5, size=N)

x = np.random.normal(10, 1, N)
y_real = alfa_real + beta_real * x
y = y_real + eps_real

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b.')
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.plot(x, y_real, 'k')
plt.subplot(1, 2, 2)
sns.kdeplot(y)
plt.xlabel("$y$", fontsize=16)
plt.show()
```

ベイジアン線形回帰モデルのパラメータのKDEとトレースを出力する

```python
with pm.Model() as model:
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=1)
    epsilon = pm.HalfCauchy('epsilon', 5)
    
    mu = pm.Deterministic('mu', alpha + beta * x)
    y_pred = pm.Normal('y_pred', mu=mu, sd=epsilon, observed=y)
    
    start = pm.find_MAP()
    step = pm.Metropolis()
    trace = pm.sample(11000, step, start, chains=1)
    
trace_a = trace[1000:]
pm.traceplot(trace_a)
```

α、βのサンプリングがうまくいっていなさそうだということが読み取れる。<br>
各パラメータの自己相関を確認してみる。

```python
varnames = ['alpha', 'beta', 'epsilon']
pm.autocorrplot(trace_a, varnames)
```

パラメータαとβの事後分布のKDEを出力する

```python
sns.kdeplot(trace_a['alpha'], trace_a['beta'])
plt.xlabel(r'$\alpha$', fontsize=16)
plt.ylabel(r'$\beta$', fontsize=16, rotation=0)
plt.show()
```

<h4>解決策⑴：中心化による解決</h4>

自己相関に問題があると、サンプリングの際に取得した値のほとんどが、確率の高い領域から外れるという問題が発生する。<br>
これを解消するために、x′=x−x¯
とデータベクトルを中心化することがあげられる。この際，各パラメータは

α=α′−βx¯，β=β′

と変換される。

<h4>解決策⑵：サンプリング方法の変更</h4>

NUTSを利用する。<br>
NUTSは、事後分布の曲率に沿って移動し、上記のplotに表されている対角線状の空間に沿って移動することが容易になる。<br>
これより以下のセルでは、NUTSを利用したサンプリングを行うこととする。

試しに上記のデータでサンプリング方法を変更した結果は下記の通り。<br>
いい感じに自己相関も減少している。

```python
with pm.Model() as model:
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=1)
    epsilon = pm.HalfCauchy('epsilon', 5)
    
    mu = pm.Deterministic('mu', alpha + beta * x)
    y_pred = pm.Normal('y_pred', mu=mu, sd=epsilon, observed=y)
    
    start = pm.find_MAP()
    step = pm.NUTS()
    trace = pm.sample(11000, step, start, chains=1)
    
trace_a = trace[1000:]
pm.traceplot(trace_a)
```

```python
varnames = ['alpha', 'beta', 'epsilon']
pm.autocorrplot(trace_a, varnames)
```

<h4>ピアソンの相関係数の推定</h4>

方法⑴：線形回帰モデルにおける回帰係数、相関係数、決定係数を出力する

```python
with pm.Model() as model:
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=1)
    epsilon = pm.HalfCauchy('epsilon', 5)
    
    mu = alpha + beta * x
    y_pred = pm.Normal('y_pred', mu=mu, sd=epsilon, observed=y)
    
    rb = pm.Deterministic('rb', (beta * x.std() / y.std()) ** 2)
    
    y_mean = y.mean()
    ss_reg = pm.math.sum((mu - y_mean) ** 2)
    ss_tot = pm.math.sum((y - y_mean) ** 2)
    rss = pm.Deterministic('rss', ss_reg/ss_tot)
    
    start = pm.find_MAP()
    step = pm.NUTS()
    trace_n = pm.sample(2000, step, start, chains=1)
    
pm.traceplot(trace_n)
```

回帰係数、相関係数の２乗、決定係数の要約統計量を出力する

```python
varnames = ['alpha', 'beta', 'epsilon', 'rb', 'rss']
pm.summary(trace_n, varnames)
```

方法⑵：多変量正規分布の共分散行列を推定する

様々な相関係数のもとで2変量正規分布を出力する

```python
sigma_x1 = 1
sigmas_x2 = [1, 2]
rhos = [-0.99, -0.5, 0, 0.5, 0.99]

x, y = np.mgrid[-5:5:0.1, -5:5:0.1]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y

f, ax = plt.subplots(len(sigmas_x2), len(rhos), sharex=True, sharey=True)

for i in range(2):
    for j in range(5):
        sigma_x2 = sigmas_x2[i]
        rho = rhos[j]
        cov = [[sigma_x1**2, sigma_x1*sigma_x2*rho], [sigma_x1*sigma_x2*rho, sigma_x2**2]]
        rv = stats.multivariate_normal([0, 0], cov)
        ax[i, j].contour(x, y, rv.pdf(pos))
        ax[i, j].plot(0, 0, label="$\\sigma_{{x2}}$ = {:3.2f}\n$\\rho$ = {:3.2f}".format(sigma_x2, rho), alpha=0)
        ax[i, j].legend()
    ax[1, 2].set_xlabel("$x_1$")
    ax[1, 0].set_ylabel("$x_2$")
    plt.tight_layout()
```
